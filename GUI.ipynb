```python
ans = 0
class_num = 9
```

```python
import tkinter as tk
from PIL import Image, ImageTk

mapping = {0:'amoxicillin',1:'Fucole PARAN',2:'Magnesium Oxide',3:'Tonec',4:'TonFul',
           5:'Ulexin',6:'BHL',7:'Biperin',8:'Peiwetsu',9:'Bismuth'}

def close_window(): #######################
    window.destroy()  #######################

if ans==0:

    window = tk.Tk()
    window.title('藥物辨識')
    window.wm_attributes('-topmost',1) #視窗出現在最上層
    window.geometry("620x480+350+120") #視窗出現位置跟大小
    label = tk.Button(window,                 # 文字標示所在視窗
                     text = '(X) 藥 品 錯 誤 ! 請 再 次 確 認 !\n正確藥品(QRcode) : '+ mapping[class_num] +"\n用途 : "
                      +effectiveness_new[0:len(effectiveness_new)//2]+'\n'+effectiveness_new[len(effectiveness_new)//2:],  # 顯示文字
                     bg = 'white',         #  背景顏色
                     fg = 'red', 
                     font = ('Forte', 20),   # 字型與大小
#                      width = 80, height = 3) # 文字標示尺寸   
                     width = 41, height = 5,command=close_window) ####################
    label.pack()

    img = ImageTk.PhotoImage(Image.open("./ng.jpg").resize((310, 310), Image.ANTIALIAS))  # NG圖
    canvas = tk.Canvas(window, height=620, width=620)
    image = canvas.create_image(155, 0, anchor='n',image=img) 
    canvas.pack()

    im_Med = Image.open("./" + mapping[class_num] + ".png") #藥物圖
    im_Med = im_Med.resize((310, 310), Image.ANTIALIAS)
    img_Med_re = ImageTk.PhotoImage(im_Med)
    image_ = canvas.create_image(310+155, 0, anchor='n',image=img_Med_re)

    window.mainloop()


elif ans==1:

    window = tk.Tk()
    window.title('藥物辨識')
    window.wm_attributes('-topmost',1)
    window.geometry("600x480+350+120") #視窗出現位置跟大小
    label = tk.Button(window,                 # 文字標示所在視窗
                     text = '(O) 藥 品 正 確 !\n'+ mapping[class_num] +"\n用途 : "+effectiveness_new[0:len(effectiveness_new)//2]+'\n'+effectiveness_new[len(effectiveness_new)//2:],  # 顯示文字
                     bg = 'white',         #  背景顏色
                     fg = '#228B22', 
                     font = ('Forte', 20),   # 字型與大小
#                      width = 80, height = 3,) # 文字標示尺寸   
                     width = 40.5, height = 5,command=close_window) ####################
    label.pack()

    img = ImageTk.PhotoImage(Image.open("./ok.jpeg").resize((310, 310), Image.ANTIALIAS))  # OK圖
    canvas = tk.Canvas(window, height=620, width=620)
    image = canvas.create_image(155, 0, anchor='n',image=img) 

    im_Med = Image.open("./" + mapping[class_num] + ".png") #藥物圖
    im_Med = im_Med.resize((310, 310), Image.ANTIALIAS)
    img_Med_re = ImageTk.PhotoImage(im_Med)
    image_ = canvas.create_image(310+155, 0, anchor='n',image=img_Med_re)

    canvas.pack()

    window.mainloop()


else:

    window = tk.Tk()
    window.title('藥物辨識')
    window.wm_attributes('-topmost',1)
    window.geometry("256x40+350+150") #視窗出現位置跟大小
    label = tk.Label(window,                 # 文字標示所在視窗
                     text = '非 預 期 錯 誤 !',  # 顯示文字
                     bg = '#F0E68C',         #  背景顏色
                     fg = '#C71585', 
                     font = ('Forte', 20),   # 字型與大小
                     width = 15, height = 2) # 文字標示尺寸   
#                      width = 15, height = 2,command=close_window) ####################
    label.pack()

    window.mainloop()
```
